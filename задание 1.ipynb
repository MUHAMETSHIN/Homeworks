```python
##Замени текстовик geiger_counter_data.txt на свой с таким же названием

import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd

detections_per_second = np.loadtxt('geiger_counter_data.txt', comments = '#', encoding = 'utf8')
detections_quantity = len(detections_per_second)

def get_detections_per_pile (pile_size):
    detections_per_pile = []
    for i in range (0, len(detections_per_second)//pile_size):
        detections_per_pile.append(0)
        for j in range(0, pile_size):
            detections_per_pile[i] += int(detections_per_second[i*pile_size + j])
        j=0
    return detections_per_pile

def get_detections_table (pile_size):
        data_table = pd.DataFrame(data = np.array(get_detections_per_pile(pile_size)).reshape(detections_quantity//pile_size//10, 10), dtype = 'int64', index = [i for i in range(0,detections_quantity//pile_size - 9,10)], columns = [c for c in range(1,11)])
        data_table.columns.name = '№ опыта'
        print("\nЧисло срабатывания счётчика за {0} c\n".center(50).format(str(pile_size)))
        display(data_table)

def get_histo_table (pile_size, border_left, border_right):
        histo_data = {'Число импульсов n': [x for x in range (border_left, border_right)], 
                      'Число случаев': [get_detections_per_pile(pile_size).count(x) for x in range (border_left, border_right)],
                      'Доля случаев ω': [get_detections_per_pile(pile_size).count(x)/(detections_quantity//pile_size) for x in range (border_left, border_right)]
                     }
        a = pd.DataFrame(histo_data, dtype = 'float64').transpose()
        a.columns = [None] * len(a.columns)
        display(a)

get_detections_table(10)
get_detections_table(40)
get_detections_table(80)

get_histo_table(10, 0, 6)
get_histo_table(10,6,12)
get_histo_table(10,12,18)

get_histo_table(40, 25, 34)
get_histo_table(40,34,43)
get_histo_table(40,43,52)

plt.style.use('classic')
#plt.hist([get_detections_per_pile(pile_size_1), get_detections_per_pile(pile_size_2)], density = True, bins = 50,stacked = True)
plt.hist(get_detections_per_pile(10), density = True, stacked = True, label = 'Срабатывание счётчика за 10с', bins = 20)
plt.hist(get_detections_per_pile(40), density = True, stacked = True, label = 'Срабатывание счётчика за 40с', bins = 20)
plt.hist(get_detections_per_pile(80), density = True, stacked = True, label = 'Срабатывание счётчика за 80с', bins = 20)
plt.xlabel('Срабатывания счётчика за секунду')
plt.ylabel('Доля случаев фиксации')
plt.title('Гистограммы распределения числа отсчетов')
#273 68.25
#387 96.75

plt.legend(loc = 'best')
```
